```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
!jupyter nbextension enable --py widgetsnbextension

```

```python
def Canny_Edge_Detector(img_input, img_output, quality, lower_threshold, upper_threshold, aperture_size, L2gradient):

    img = cv2.imread(img_input,0)
    edges = cv2.Canny(img,lower_threshold, upper_threshold, apertureSize=aperture_size, L2gradient=L2gradient)
    cv2.imwrite(img_output, img)
    _=plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

    plt.show()
```

```python
interact(Canny_Edge_Detector, img_input='denoise.png', img_output='edge_detection'+'.jpg',\
        quality=50,\
         lower_threshold=(0,255,1), upper_threshold=(0, 255, 1),\
        aperture_size=fixed(3), L2gradient=False);
```
